```python
import tensorflow as tf
from tensorflow import keras
import numpy as np
import matplotlib.pyplot as plt
```

**1. Load and preprocess data**

```python
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```python
print('x train size= ',len(x_train))
print('y train size= ',len(y_train))
print('x test size= ',len(x_test))
print('y test size= ',len(y_test))
```

```python
x_train, x_test = x_train / 255.0, x_test / 255.0
```

**2. Create validation set**

```python
x_val = x_train[55000:60000]
x_train = x_train[0:55000]
len(x_val)
```

```python
y_val=y_train[55000:60000]
y_train=y_train[0:55000]
print("\n",len(y_val))
```

**3. Create model**

```python
model= keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28), name='Flatten_input'),
    keras.layers.Dense(128,activation="relu", name='Hidden_dense_layer'),
    keras.layers.Dense(10,activation="softmax", name='Output_dense_layer')])
```

**4. Compile Model**

```python
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'])
```

**5. Train WITHOUT EarlyStopping**

```python
history=model.fit(x_train,y_train,epochs=20,validation_data=(x_val,y_val))
```

**6. Plot curves**

```python
plt.plot(history.history['loss'],label='Loss')
plt.plot(history.history['val_loss'],label='Validation Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')
plt.show()
```

```python
model.evaluate(x_test,y_test)
```

# Task 1 — Deep Prediction Analysis

```python
sample1_x = x_test[300]
sample1_y = y_test[300]

sample2_x = x_test[500]
sample2_y = y_test[500]

sample3_x = x_test[38]
sample3_y = y_test[38]

samples = [
    ("Sample 1", sample1_x, sample1_y),
    ("Sample 2", sample2_x, sample2_y),
    ("Sample 3", sample3_x, sample3_y)]

for name, sample_x, sample_y in samples:

    # Prediction
    pred = model.predict( sample_x.reshape(1,28,28),verbose=0)
    predicted_label = np.argmax(pred)
    confidence = np.max(pred)

    # Display image
    plt.figure(figsize=(3,3))
    plt.imshow(sample_x, cmap='gray')
    plt.title(name)
    plt.axis('off')
    plt.show()

    # Print results
    print("="*60)
    print(name)
    print("="*60)

    print(f"True Label      : {sample_y}")
    print(f"Predicted Label : {predicted_label}")
    print("\nProbability Vector:")
    print(np.round(pred[0],4))
    print(f"Confidence Score: {confidence:.4f}")

    print()
```

- The network is saying:
"I am 99.44% sure this image is a 3."
But the true label from the dataset is:2

So the model is very confident and very wrong.


- The **Softmax layer** only measures:


How strongly the network prefers one class over the others
It does NOT know whether the answer is actually correct.
The model does not see the true label during prediction.
It only computes:

Input image

    ↓
Hidden features


    ↓

Class scores

    ↓

Softmax probabilities

and chooses the largest probability.

### Interpretation :

1.    The Input image is first passed through the **Flatten layer**, which converts the 28×28 image into a 1D vector. This vector is then processed by the **hidden layer**, where each neuron computes weighted combinations of the input features.

2. **The ReLU activation function** is applied in the hidden layer to introduce non-linearity and help the network learn important patterns such as edges, curves, and digit shapes. ReLU also improves training efficiency by reducing vanishing gradient problems.

3. **The Output layer uses the Softmax**  activation function, which converts the network outputs into probabilities for the ten digit classes (0–9). The class with the highest probability is selected as the predicted digit.

4. **Adam Optimizer:**  During training, the Adam optimizer adjusted the network weights by minimizing the prediction error. Adam uses adaptive learning rates, which helps the model converge faster and achieve high classification accuracy.

# Task 2 — Custom Image Generalization Test

```python
from google.colab import files
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
import cv2

# Upload 3 images
print("Please upload 3 images (e.g., 1.png, 2.png, 3.png)")
uploaded = files.upload()

# Limit to the first 3 if more are uploaded, or process what's available
file_names = list(uploaded.keys())[:3]
```

```python
for file_name in file_names:
    print(f"\n--- Processing {file_name} ---")

    # 1. Open and preprocess
    img = Image.open(file_name)
    img_gray = img.convert("L")
    img_inverted = ImageOps.invert(img_gray)
    img_np = np.array(img_inverted)


# 2. Apply Dilation to thicken lines
    kernel = np.ones((5, 5), np.uint8)
    dilated_np = cv2.dilate(img_np, kernel, iterations=1)

    # 3. Center the digit as MNIST Image
    coords = cv2.findNonZero(dilated_np)
    if coords is not None:
        x, y, w, h = cv2.boundingRect(coords)
        digit_crop = dilated_np[y:y+h, x:x+w]

        canvas_size = 28
        ratio = 20.0 / max(w, h)
        new_w, new_h = max(1, int(w * ratio)), max(1, int(h * ratio))
        digit_resized = cv2.resize(digit_crop, (new_w, new_h), interpolation=cv2.INTER_AREA)

     # Place it in the center of the 28x28 background
        final_img = np.zeros((canvas_size, canvas_size), dtype=np.uint8)
        offset_x, offset_y = (canvas_size - new_w) // 2, (canvas_size - new_h) // 2
        final_img[offset_y:offset_y+new_h, offset_x:offset_x+new_w] = digit_resized

      # 4. Normalize and Predict
        img_normal = final_img.astype('float32') / 255.0
        input_data = img_normal.reshape(1, 28, 28)

        pred = model.predict(input_data, verbose=0)
        predicted_label = np.argmax(pred)
        confidence = np.max(pred)

        # 5. Display Result
        fig, axes = plt.subplots(1, 2, figsize=(8, 3))
        axes[0].imshow(img)
        axes[0].set_title(f"Original: {file_name}")
        axes[0].axis("off")

        axes[1].imshow(img_normal, cmap='gray')
        axes[1].set_title(f"Pred: {predicted_label} ({confidence:.2%})")
        axes[1].axis("off")
        plt.show()
    else:
        print(f"Could not detect any digit in {file_name}.")
```

### **Conclusion**

After testing numerous handwritten numbers, it became clear that the model is very accurate but requires several preliminary operations to make correct predictions. These essential steps include:

1.  **Grayscale & Invert**: Convert to grayscale and ensure the digit is white on a black background.
2.  **Dilation**: Use a kernel to thicken the handwritten strokes.
3.  **Resize & Center**: Fit the digit into a 20x20 area centered within a 28x28 canvas.
4.  **Normalize**: Scale pixel values to the range [0, 1].
5.  **Reshape**: Prepare the array shape to `(1, 28, 28)` for the model input layer.

# Task 3 — Epoch-Based Learning Curve Exploration

```python
# definition function take parameter epoce_count and return history

def train_and_plot(epoche_count):
    print(f"\n{'='*20} Training for {epoche_count} Epochs {'='*20}")

    # Re initialize temp_model to ensure a fresh start
    temp_model= keras.Sequential([
        keras.layers.Flatten(input_shape=(28,28)),
        keras.layers.Dense(128, activation='relu'),
        keras.layers.Dense(10, activation='softmax')
    ])
    temp_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    history = temp_model.fit(x_train, y_train, epochs= epoche_count, validation_data=(x_val, y_val), verbose=0)

    #ploting
    fig,(ax1,ax2) =plt.subplots(1,2,figsize=(15,5))
    # 1- Loss Plot
    ax1.plot(history.history['loss'], label='Train Loss')
    ax1.plot(history.history['val_loss'], label='Val Loss')
    ax1.set_title(f'Loss: {epoche_count} Epochs')
    ax1.set_xlabel('Epoch')
    ax1.set_ylabel('Loss')
    ax1.legend()

      # 2- Accuracy Plot
    ax2.plot(history.history['accuracy'], label='Train Acc')
    ax2.plot(history.history['val_accuracy'], label='Val Acc')
    ax2.set_title(f'Accuracy: {epoche_count} Epochs')
    ax2.set_xlabel('Epoch')
    ax2.set_ylabel('Accuracy')
    ax2.legend()

    plt.show()
    return history
# Now Excecute for 5,10,20 epochs
hist5= train_and_plot(5)

```

```python
hist10= train_and_plot(10)
```

```python
hist20= train_and_plot(20)
```

### **Analysis of Results**

*   **At 5 Epochs:** The model is in the 'Learning' phase. Both training loss and validation loss are decreasing steadily. There are **no signs of overfitting** yet; the model is generalizing well.
*   **At 10 Epochs:** The validation loss starts to flatten out (plateau). While training accuracy continues to climb, the gap between training and validation accuracy begins to widen slightly.
*   **At 20 Epochs:** **Clear signs of overfitting appear.** we will notice that the **Training Loss** continues to decrease towards zero, but the **Validation Loss** starts to increase or fluctuate. This validation loss curve indicates that the model is no longer learning general patterns of digits but is instead 'memorizing' the specific noise and details of the training images.


# Task 4 — EarlyStopping Behavior Analysis

```python
model_es = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(64, activation='relu'),
    keras.layers.Dense(10, activation='softmax')
])
```

```python
model_es.compile( optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
callback= tf.keras.callbacks.EarlyStopping(patience=3,restore_best_weights=True,monitor= 'val_loss' )
history=model_es.fit(x_train,y_train,epochs=20,validation_data=(x_val,y_val),callbacks=[callback])
```

### **Interpreter: EarlyStopping Result**

*   **Automatic Stop**: The model did not reach all 20 epochs; it stopped at **Epoch 10** because it detected that the validation loss was no longer decreasing.
*   **Patience in Action**: Since `patience=3`, the model saw that the loss at Epochs 8, 9, and 10 was worse than at Epoch 7, so it ended the training to prevent overfitting.
*   **With Patience = 5**: The training is more flexible. It waits longer (5 epochs) to see if the loss will drop again. This can lead to slightly higher accuracy but might allow the training to run for unnecessary extra cycles if the model has already peaked.
*   **Best Weights**: Thanks to `restore_best_weights=True`, the model you have now is the one from **Epoch 7**, ensuring you have the most accurate version for future predictions.

# Task 7 — Optimizer Comparison Challenge

```python
def build_base_model():
    return keras.Sequential([keras.layers.Flatten(input_shape=(28, 28), name='Flatten_input'),
                            keras.layers.Dense(128, activation='relu', name='Hidden_dense_layer'),
                            keras.layers.Dense(10, activation='softmax', name='Output_dense_layer')
                            ])
```

```python
optimizers_dict = {
    'SGD':            keras.optimizers.SGD(learning_rate=0.01),
    'SGD + Momentum': keras.optimizers.SGD(learning_rate=0.01, momentum=0.9),
    'Adam':           keras.optimizers.Adam(learning_rate=0.001),
    'AdamW':          keras.optimizers.AdamW(learning_rate=0.001, weight_decay=0.004),
}

Epochs = 10
histories_opt = {}

for name, opt in optimizers_dict.items():
    print(f"\n{'='*20} Training with {name} {'='*20}")
    m = build_base_model()
    m.compile(optimizer=opt, loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    hist = m.fit(x_train, y_train, epochs=Epochs,validation_data=(x_val, y_val), verbose=0,batch_size= 64 )
    histories_opt[name] = hist
```

```python
# Individual loss/accuracy curves per optimizer
for name, hist in histories_opt.items():
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 4))

    ax1.plot(hist.history['loss'], label='Train Loss')
    ax1.plot(hist.history['val_loss'], label='Val Loss')
    ax1.set_title(f'{name} — Loss')
    ax1.set_xlabel('Epoch'); ax1.set_ylabel('Loss'); ax1.legend()

    ax2.plot(hist.history['accuracy'], label='Train Acc')
    ax2.plot(hist.history['val_accuracy'], label='Val Acc')
    ax2.set_title(f'{name} — Accuracy')
    ax2.set_xlabel('Epoch'); ax2.set_ylabel('Accuracy'); ax2.legend()

    plt.show()
```

```python
for name, hist in histories_opt.items():
    print(f"{name:16s} -> final val_loss={hist.history['val_loss'][-1]:.4f}, "
          f"final val_accuracy={hist.history['val_accuracy'][-1]:.4f}")
```

### **Interpretation of Optimizer Results**

1.  **SGD**: Basic and slow; it struggles to reach peak performance within 10 epochs.
2.  **SGD + Momentum**: significantly better than basic SGD; the 'momentum' helps it avoid getting stuck and speeds up learning.
3.  **Adam**: Very reliable and fast, consistently providing high accuracy.
4.  **AdamW**: The top performer here. By adding 'Weight Decay' to the Adam algorithm, it generalizes better to the validation set, achieving the lowest error and highest accuracy overall.

# Task 8 — Batch Size & Gradient Noise Experiment

```python
batch_sizes = [8, 32, 128]
epochs= 10
histories= {}

for batch in batch_sizes:
    print(f"\n{'='*20} Training with batch_size={batch} {'='*10}")
    mod = build_base_model()
    mod.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

    hist = mod.fit(x_train, y_train,epochs=epochs,batch_size=batch,  validation_data=(x_val, y_val),verbose=0)
    histories[batch] = hist
```

```python
colors_bs = {8: 'crimson', 32: 'seagreen', 128: 'steelblue'}

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
for bs, hist in histories.items():
    ax1.plot(hist.history['loss'], color=colors_bs[bs], label=f'batch={bs} (train)')
    ax1.plot(hist.history['val_loss'], '--', color=colors_bs[bs], alpha=0.6)
    ax2.plot(hist.history['accuracy'], color=colors_bs[bs], label=f'batch={bs} (train)')
    ax2.plot(hist.history['val_accuracy'], '--', color=colors_bs[bs], alpha=0.6)

ax1.set_title('Loss vs Batch Size'); ax1.set_xlabel('Epoch'); ax1.set_ylabel('Loss'); ax1.legend()
ax2.set_title('Accuracy vs Batch Size'); ax2.set_xlabel('Epoch'); ax2.set_ylabel('Accuracy'); ax2.legend()
plt.show()

for bs, hist in histories.items():
    print(f"batch_size={bs:4d} -> final val_loss={hist.history['val_loss'][-1]:.4f}, "
          f"final val_accuracy={hist.history['val_accuracy'][-1]:.4f}")
```

### **Interpretation of Batch Size Results**

1.  **Small Batch (batch=8):** Provides the most 'noise' during training. While it can help the model escape local minima, it often leads to less stable loss curves and slightly higher validation loss if the noise is too high.
2.  **Medium Batch (batch=32):** A common standard that balances training speed and gradient stability.
3.  **Large Batch (batch=128):** In this specific run, the **large batch size (128)** achieved the **lowest validation loss**. This suggests that for the MNIST task, smoother gradient updates helped the model find a more stable and generalizable solution.

**Conclusion:** Larger batch sizes often lead to faster training and smoother convergence, while smaller batches can sometimes improve generalization through 'gradient noise,' though they are computationally slower per epoch.


# Task 9 — Activation Function Swap (ReLU vs Tanh vs GELU)

- Same architecture and optimizer, only the hidden-layer activation changes.

```python
Activ=['relu','tanh','softsign','gelu']
epochs=10
histories_act= {}

def model_with_activ(activation_name):
    return keras.Sequential([keras.layers.Flatten(input_shape=(28, 28), name='Flatten_input'),
                            keras.layers.Dense(128, activation=activation_name, name='Hidden_dense_layer'),
                            keras.layers.Dense(10, activation='softmax', name='Output_dense_layer')
                            ])
```

```python
for actv in Activ:
  print(f"\n{'='*20} Training with {actv} activation {'='*20}")

  mod_actv= model_with_activ(actv)
  mod_actv.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
  hist_actv = mod_actv.fit(x_train, y_train,epochs=epochs,validation_data=(x_val, y_val), verbose=0)

  histories_act[actv] = hist_actv
```

```python
colors_act = {'relu': 'green', 'tanh': 'purple', 'softsign': 'darkorange', 'gelu': 'royalblue'}

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))
for act, hist in histories_act.items():
    ax1.plot(hist.history['loss'], color=colors_act[act], label=f'{act} (train)')
    ax1.plot(hist.history['val_loss'], '--', color=colors_act[act], alpha=0.6)
    ax2.plot(hist.history['accuracy'], color=colors_act[act], label=f'{act} (train)')
    ax2.plot(hist.history['val_accuracy'], '--', color=colors_act[act], alpha=0.6)

ax1.set_title('Loss vs Activation Function'); ax1.set_xlabel('Epoch'); ax1.set_ylabel('Loss'); ax1.legend()
ax2.set_title('Accuracy vs Activation Function'); ax2.set_xlabel('Epoch'); ax2.set_ylabel('Accuracy'); ax2.legend()
plt.show()

for act, hist in histories_act.items():
    print(f"{act:10s} -> final val_loss={hist.history['val_loss'][-1]:.4f}, "
          f"final val_accuracy={hist.history['val_accuracy'][-1]:.4f}")
```

### **Why didn't ReLU 'win' this time?**

- **GELU** achieved the highest accuracy, and **Softsign** achieved the lowest validation loss.

1.  **GELU (Gaussian Error Linear Unit):** This is a 'smoother' version of ReLU. While ReLU suddenly 'dies' at zero, GELU has a small curvature. This smoothness often helps the model learn more complex patterns and generalize better on validation data.
2.  **The 'Dying ReLU' Problem:** Sometimes, ReLU neurons can get 'stuck' at zero during training and stop learning entirely. Softsign and GELU allow for small gradients even for negative values, which keeps the learning process more active.
3.  **Randomness in Training:** Neural networks start with random weights. In a small task like MNIST, these random starts can sometimes give a slight edge to one activation function over another, even if ReLU is generally excellent.

**Summary:** ReLU is the fastest, but **GELU** and **Softsign** can sometimes find 'smoother' solutions that lead to higher precision and lower error.
